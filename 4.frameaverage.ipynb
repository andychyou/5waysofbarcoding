```python
def getColor(frame):
    height, width, depth = frame.shape
    frame_colors = []
    for x in range(int(width)):
        for y in range(int(height)):
        
            color = frame[y,x]
            frame_colors.append(color)
     
   # print(frame_colors)
    return frame_colors
```

```python
def avrg(alist):
    red_total = 0 
    green_total = 0
    blue_total = 0
    for x in alist:
        #print(x)
        red = x[0] ; green = x[1] ; blue = x[2]
        #print(red,green,blue)
        red_total += red ; green_total += green ; blue_total += blue
    red = red_total // len(alist)
    green = green_total // len(alist)
    blue = blue_total // len(alist)
    colors = [red, green, blue]
    print(colors)
    return colors
```

```python
filename='C:\\Users\\Hanrabong\\DESKTOP\\highlight.mp4'
cap = cv2.VideoCapture(filename) 
CList = []
```

```python
while(True): 
    #print('hi')
    ret, frame = cap.read()
    #print('suc')
    if ret == False:
        break 
    
  
    CList.append(avrg(getColor(frame)))
```

```python
cap.release()
barcode = np.zeros((1000,len(CList),3)) 
```

```python
for i in range(len(CList)):
    barcode[:,i] = CList[i]
```

```python
b, g, r = cv2.split(barcode)   
barcode2 = cv2.merge([r,g,b]) 
barcode2 = barcode2.astype(np.uint8) 
```

```python
plt.imshow(barcode)
plt.show()

imageio.imwrite('highlight_frameavg.png', barcode)

cv2.destroyAllWindows()
```
